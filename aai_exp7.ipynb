```python
class VirtualEnvironment:
    def __init__(self, size=5):
        # Create a grid-based environment of a given size
        self.size = size
        self.grid = [[' ' for _ in range(size)] for _ in range(size)]
        self.agent_position = [0, 0]  # Agent starts at the top-left corner
        self.grid[0][0] = 'A'  # Mark agent's initial position
        self.objects = [(2, 2), (1, 3), (3, 1)]  # Predefined object locations
        for obj in self.objects:
            self.grid[obj[0]][obj[1]] = 'O'  # Place objects in the grid

    def display(self):
        # Display the environment (grid)
        print()
        for row in self.grid:
            row_display = ""
            for cell in row:
                if cell == 'A':
                    row_display += " 🟢 A "  # Green circle emoji for agent
                elif cell == 'O':
                    row_display += " 🔴 O "  # Red circle emoji for object
                else:
                    row_display += "     "  # Empty space
            print(row_display)
        print()

    def move_agent(self, direction):
        # Move agent based on input direction (up, down, left, right)
        current_x, current_y = self.agent_position
        if direction == 'up' and current_x > 0:
            self.agent_position = [current_x - 1, current_y]
        elif direction == 'down' and current_x < self.size - 1:
            self.agent_position = [current_x + 1, current_y]
        elif direction == 'left' and current_y > 0:
            self.agent_position = [current_x, current_y - 1]
        elif direction == 'right' and current_y < self.size - 1:
            self.agent_position = [current_x, current_y + 1]
        self.update_grid()

    def update_grid(self):
        # Update the grid with the agent's new position
        self.grid = [[' ' for _ in range(self.size)] for _ in range(self.size)]
        self.grid[self.agent_position[0]][self.agent_position[1]] = 'A'
        for obj in self.objects:
            self.grid[obj[0]][obj[1]] = 'O'

    def check_for_interaction(self):
        # Check if the agent interacts with any object (if they are at the same position)
        if tuple(self.agent_position) in self.objects:
            return "🎯 Interaction with object detected!"  # Target emoji for interaction
        return "❌ No interaction."

# Main simulation
if __name__ == "__main__":
    env = VirtualEnvironment(size=5)

    # Instructions for the user
    print("🌟 Welcome to the stylish interactive simulation! 🌟")
    print("🎮 You can control the agent with the following commands:")
    print("'up' - Move up, 'down' - Move down, 'left' - Move left, 'right' - Move right.")
    print("🔴 The agent will interact with objects (denoted by 'O') when they reach them.")
    print("💬 Type 'exit' to quit the game.\n")

    while True:
        env.display()
        # Get user input for direction
        direction = input("🕹️ Enter direction (up/down/left/right): ").lower()

        if direction == 'exit':
            print("👋 Exiting the simulation. Thank you for playing! 🌟")
            break

        if direction in ['up', 'down', 'left', 'right']:
            env.move_agent(direction)
            interaction = env.check_for_interaction()
            print(interaction)
        else:
            print("⚠️ Invalid direction. Please enter one of 'up', 'down', 'left', or 'right'.")

        print("───────────────────────────────────────────")
```
